Para simplificar la primera versión de la librería, el responsable del proyecto decidió que todas las personas están asociadas a la misma categoría y que una lista que maneja cada centro indica los pacientes que pueden pasar al otro centro y aquellos pacientes que no están en dicha lista son los que preferiría tratar el centro.</p>

De esta manera deben crear una función "<b>puedecambiar</b>" que reciba la lista de los índices de los pacientes que están asignados al centro secundario y la lista de los índices de los pacientes asignados al centro primario y que retorne la cantidad máxima de pacientes que pueden cambiar, teniendo en cuenta que el cambio uno a uno (un paciente del centro primario por uno del secundario).</b>

```python
# Method that generates the categories of patients
def categorias(patient_categories):
  categories = []

  for category in patient_categories:
    if not category in categories:
      categories.append(category)
  
  return categories

def mefaltandelacategoria(unassigned_patients,patients_to_treat,category):
  assing_patients = []

  for patient in unassigned_patients:
    if patients_to_treat[patient] == category:
      assing_patients.append(patient)
  
  return assing_patients

def notengopacientes(secondary_center,primary_center):
  patient_to_treat = []

  for patient in secondary_center:
    if not patient in primary_center:
      patient_to_treat.append(patient)

  return patient_to_treat

def puedecambiar(secondary_center,primary_center):
  secondary_center_change = []
  primary_center_change = []

  for patient in secondary_center:
    if not patient in primary_center:
      primary_center_change.append(patient)

  for patient in primary_center:
    if not patient in secondary_center:
      secondary_center_change.append(patient)

  return len(primary_center_change) if len(primary_center_change) < len(secondary_center_change) else len(secondary_center_change) 

print(categorias(['B','B','E','C','C','E','J']))
print(mefaltandelacategoria([1,3,6,8], ['E','B','C','A','D','E','B','D','A','F'], 'B'))
print(notengopacientes ([3,5,7,10,15,16],[4,10,5,8]))
print(puedecambiar([10, 6, 9, 3, 4, 8],[ 11, 6, 8, 12, 13, 9, 5]))
```
